```python
import matplotlib.pyplot as plt
import random
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

# 图像基本绘制操作

```python
#创建画布
#plt.figure(figsize=(20,10))
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(20,10),dpi =0.7)

#创建数据（将两个折线画在同一画布）
x = range(60)
y_beijing =[random.uniform(10,15) for i in x] 
y_shanghai =[random.uniform(15,20) for i in x] 

#制作折线
#plt.plot(x,y_beijing,label = '北京',linestyle = '-.')
#plt.plot(x,y_shanghai,label = '上海')
axes[0].plot(x,y_beijing,label = '北京',linestyle = '-.')
axes[1].plot(x,y_shanghai,label = '上海')

#图例设置
#plt.legend()  #legend(loc = 'best'),loc可选择0-10的数字或者best来确定图例位置
axes[0].legend(loc = 0)
axes[1].legend(loc = 2)
#x轴y轴的刻度
x_ticks_labels = ["11点{}分".format(i) for i in x]   #x轴的标签
y_ticks = range(40)
#plt.xticks(x[::5],x_ticks_labels[::5]) #该函数第一个传入的是数字
#plt.yticks(y_ticks)
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks)
axes[0].set_xticklabels(x_ticks_labels[::5])
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks)
axes[1].set_xticklabels(x_ticks_labels[::5])


#添加x轴，y轴以及图表的标签
#plt.xlabel('时间')
#plt.ylabel('温度')  
#plt.title('中午11点到12点的温度变化')
axes[0].set_xlabel('时间')
axes[0].set_ylabel('温度')
axes[0].set_title('北京中午11点到12点的温度变化')
axes[1].set_xlabel('时间')
axes[1].set_ylabel('温度')
axes[1].set_title('上海中午11点到12点的温度变化')

#网格设置
#plt.grid(linestyle = '--',alpha = 0.7)  
axes[0].grid(linestyle = '--',alpha = 0.7)
axes[1].grid(linestyle = '--',alpha = 0.7)

plt.show()
```

```python
x = np.linspace(-10,10,1000)
y = np.sin(x)

plt.figure(figsize=(20,10),dpi = 100)

plt.plot(x,y,linestyle = '-.')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.title('y = sin(x)')
plt.grid(linestyle= '-',alpha= 0.7)

plt.show()
```
