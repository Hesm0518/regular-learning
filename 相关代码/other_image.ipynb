```python
#获取数据
x = [22,33,44,55,66,77,88,88,99,55,6,77,87,65]
y= [11,13,23,34,45,56,67,78,89,67,78,77,67,63]

#创建画布
plt.figure(figsize=(10,10),dpi=100)

#绘制图像
plt.scatter(x,y)

#展示图像
plt.show()
```

#柱状图

```python
#获取数据
movie_name = ['雷神3','流浪地球','战狼2','宝贝计划','美国队长','变形金刚3','独行月球','怦然心动','泰坦尼克号','霸王别姬','冰雪奇缘','飞行屋','蜘蛛侠']
y = [10000,14423,24545,35453,23434,10034,13020,46567,35455,60332,45458,24344,7890]
x = range(len(movie_name))
#创建画布
plt.figure(figsize=(15,9),dpi=100)

#绘制图像
plt.bar(x, y, color = ['r','b','g','y','c','r','k','m','g','r','c','b','y'], width = 0.5)


plt.xticks(x, movie_name, fontsize = 10)
plt.xlabel('电影名')
plt.ylabel('电影票房数')
plt.grid()
plt.title('某些电影票房')
plt.savefig('./bartest.png')
#展示图像
plt.show()
```
